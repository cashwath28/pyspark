```
data = [(1, 'Chandra', 'Mobile'), (1, 'Chandra', 'TV'), (1, 'Chandra', 'Monitor'), (2, 'Shekar', 'Mobile'), (2, 'Shekar', 'TV'), (2, 'Shekar', 'Monitor')]
columns = ('CustomerID', 'Name', 'Purchase')

df = spark.createDataFrame(data, columns)
df.show()
```

```
df.createOrReplaceTempView("Sample")
```

```
%sql
select CustomerID, Name, collect_list(Purchase) from Sample
group by 1, 2
```

```
from pyspark.sql.functions import collect_list
df1 = df.groupBy('CustomerID', 'Name').agg(collect_list('Purchase'))
df1.show()
```
